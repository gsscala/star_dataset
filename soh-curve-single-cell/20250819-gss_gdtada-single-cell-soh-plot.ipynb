# Battery State of Health (SOH) Analysis and Visualization

Visualizes and analyzes the state of health (SOH) of battery cells. It uses capacity data to generate plots that help us understand the cell's behavior throughout its lifespan.

The expected result was to plot a graph showing the SoH decreasing over time, and this was achieved.

To run the notebook, you need to have the CU_Dynamic folder in this directory.

```python
from soh_plot_functions import *
```

```python
# Load and analyze initial check-up data for cell BW-VTC-458
initial_data = load_cell_data("../CU_Dynamic/CU000/BW-VTC-458_7068_CU_cyc_000_BW-VTC-AgeVal.mat")
plot_capacity_vs_time(initial_data["Time"], initial_data["I"])
```

```python
# Load and analyze final check-up data for cell BW-VTC-458
final_data = load_cell_data("../CU_Dynamic/CU011/BW-VTC-458_8109_CU_cyc_011_BW-VTC-AgeVal.mat")
plot_capacity_vs_time(final_data["Time"], final_data["Ah"])
```

```python
# Extract capacity data from all check-up cycles
capacity_values, checkup_cycles = extract_capacity_data("../CU_Dynamic", "BW-VTC-458")

print(f"Processed {len(checkup_cycles)} check-up cycles")
```

```python
# Calculate State of Health metrics
soh_percentages = calculate_soh(capacity_values)

# Display results
print("\nSoH Results:")
for i in range(len(checkup_cycles)):
    print(f"Check-up {checkup_cycles[i]}: {soh_percentages[i]:.1f}%")
```

```python
# Visualize battery health degradation trend
plot_soh_vs_checkups(checkup_cycles, soh_percentages)
```
