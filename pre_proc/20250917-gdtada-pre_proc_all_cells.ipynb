```python
src = "../../../../../../media/mods-pred/Datasets/Data_Munich_h5"
sufix = "_metrics"

import os

for cell in sorted(os.listdir(src)):  
    cell_path = os.path.join(src, cell)
    for checkup in sorted(os.listdir(cell_path)):
        if checkup.endswith(".h5") and checkup.startswith("CU"):
            path = os.path.join(cell_path, checkup) 
            df = load_h5(path)
            df = process_states(df)
            metrics = analyze_metrics_by_status_period(df)

            metrics_file_name = checkup.replace(".h5", sufix + ".h5")
            metrics_path = os.path.join(cell_path, metrics_file_name)
            metrics.to_hdf(metrics_path, key="metrics", mode="w", complevel=9, complib='zlib')

            print(f"Processed {metrics_path}")
```
